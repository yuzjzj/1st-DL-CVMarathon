使用 [Lena](http://optipng.sourceforge.net/pngtech/img/lena.png)

- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = 'data/lena.png'
```

```python
# 以彩色圖片的方式載入
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

# 改變不同的 color space
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
img_lab2 = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_hlsf = cv2.cvtColor(img, cv2.COLOR_BGR2HLS_FULL)

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.axis('off')
plt.imshow(img[:,:,::-1])
plt.title('BGR')
plt.subplot(2,3,2)
plt.axis('off')
plt.imshow(img_lab[:,:,::-1])
plt.title('LAB')
plt.subplot(2,3,3)
plt.axis('off')
plt.imshow(img_lab2[:,:,::-1])
plt.title('Lab')
plt.subplot(2,3,4)
plt.axis('off')
plt.imshow(img_hsv[:,:,::-1])
plt.title('HSV(HSB)')
plt.subplot(2,3,5)
plt.axis('off')
plt.imshow(img_hls[:,:,::-1])
plt.title('HLS')
plt.subplot(2,3,6)
plt.axis('off')
plt.imshow(img_hlsf[:,:,::-1])
plt.title('HLS_FULL')
plt.show()
```
